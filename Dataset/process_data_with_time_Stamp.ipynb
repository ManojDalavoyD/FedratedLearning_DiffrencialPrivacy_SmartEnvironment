```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("smart_home_energy_consumption_large (1).csv")
#order the home id in asscending order
df=df.sort_values(by=['Home ID'])
df.head()
```

```python
#merge the same home id into one row appliances into colums inside each appliance energynConsumed and time tamp respectively
grouped=df.groupby('Home ID')
merged_df=grouped.agg({
    'Time': lambda x: ','.join(x.astype(str)),
    'Appliance Type': lambda x: ','.join(x.astype(str)),
    'Energy Consumption (kWh)': lambda x: ','.join(x.astype(str)),
    'Outdoor Temperature (°C)':lambda x: ','.join(x.astype(str)),
    'Household Size': lambda x:','.join(x.astype(str)),
}).reset_index()
merged_df.head()
```

```python
merged_df.to_csv("Smart_homes_data_along_with_time_stamp.csv",index=False)
merged_df.to_excel("Smart_homes_data_along_with_time_stamp.xlsx",index=False
                   )
```
